```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
base_path = "/Users/flock/Documents/ML/Titanic"
train_df = pd.read_csv(base_path + "/train.csv")
test_df = pd.read_csv(base_path + "/test.csv")

train_data = train_df.drop(["Name", "PassengerId", "Cabin", "Ticket"], axis=1)
test_data = test_df.drop(["Name", "PassengerId", "Cabin", "Ticket"], axis=1)

train_data.head()
```

```python
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].mean())
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].mean())

test_data["Fare"] = test_data["Fare"].fillna(test_data["Fare"].mean())

train_data["Sex"] = train_data["Sex"].apply(lambda x: 1 if (x == "female") else 0 )
test_data["Sex"] = test_data["Sex"].apply(lambda x: 1 if (x == "female") else 0 )
train_data["Embarked"] = train_data["Embarked"].apply(lambda x: 0 if pd.isna(x) else (ord(x) - 64) )
test_data["Embarked"] = test_data["Embarked"].apply(lambda x: 0 if pd.isna(x) else (ord(x) - 64) )

# train_data = train_data[pd.notnull(train_data['Embarked'])]
train_data.head()
```

```python
train_data = pd.get_dummies(train_data, columns=["Embarked"])
test_data = pd.get_dummies(test_data, columns=["Embarked"])
```

```python
X = train_data.drop("Survived", axis=1)
Y = train_data["Survived"]
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
x_train.head()
```

```python
test_data.head()
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
# prediction = model.predict(test_data)
model.score(x_test, y_test)
```
